```python
#8. 
# the sample is considered a randomized sample
time_sampel = [19, 25, 22, 14, 18, 15, 17, 22, 16, 21, 
15, 20, 20, 14, 26, 14, 11, 18, 20, 16,
13, 21,  22, 15,  8, 27, 23, 20, 18, 16]
```

```python
#normality test
from scipy.stats import shapiro
alpha = 0.05
test_statistic, pvalue = shapiro(time_sampel)
print(pvalue)
if pvalue <= alpha:
    print('Reject H0: The data is not drawn from normally distributed population.')
else:
    print('Failed to reject H0: The data is drawn from normally distributed population.')
```

```python
from scipy.stats import kstest

test_statistic, pvalue = kstest(time_sampel, 'norm')
if pvalue <= alpha:
    print('Reject H0: The data is not drawn from normally distributed population.')
else:
    print('Failed to reject H0: The data is drawn from normally distributed population.')
```

```python
from statsmodels.stats.diagnostic import lilliefors

# test_statistic, pvalue = 
test_statistic, pvalue = lilliefors(time_sampel)
if pvalue <= alpha:
    print('Reject H0: The data is not drawn from normally distributed population.')
else:
    print('Failed to reject H0: The data is drawn from normally distributed population.')
```

```python
#Based from three normality tests, two out of three considered the data is drawn from a normality distribution, thus
# the hypothesis testing method for the data is T-test single population mean
    H0 < 20 The time needed for the courier to pick up the package after being asked is below 20 mins
    H1 > 20 The time needed for the courier to pick up the package after being asked is more than 20 mins of wait
```

```python
#Hypo testing for t-test single pop mean
from scipy.stats import ttest_1samp
tset, pvalue = ttest_1samp(time_sampel, 20, alternative= 'greater')
print(tset,pvalue)
if pvalue <= alpha:
    print('Reject H0: The time needed for the courier to pick up the package after being asked is more than 20 mins of wait')
else:
    print('Failed to reject H0: The time needed for the courier to pick up the package after being asked is below 20 mins of wait')
```

```python
df_order['tanggal']=pd.to_datetime(df_order['order_purchase_timestamp'])
bulan = df_order['tanggal'].dt.to_period('M')
bulan.dt.to_timestamp()
```

```python
df_order['tahun'] = df_order['tanggal'].dt.to_period('Y')
```

```python
bulantahun = df_order['tanggal'].dt.to_period('m')
bulan = []
for x in bulantahun:
    casteddate = str(x)
    splitteddate = casteddate.split('-')
    bulan.append(splitteddate[1])
df_order['bulan'] = bulan
df_order
```

```python
df_order_2017 = df_order[(df_order['tahun']=='2017')&(df_order['order_status']=='delivered')]
df_order_2017.head()
order_2017 = df_order_2017.groupby('bulan')['order_id'].nunique().reset_index().rename(columns={'order_id':'n_orders'})
order_2017
```
